Processing lyrics to dataframe

```python
file_path = 'Lyrics_Adele.json' # file path
with open(file_path) as file:
    data = json.load(file)
if 'songs' in data:
    df_songs = json_normalize(data['songs'],
                              sep='_',
                              record_path=None,
                              meta=['title', 'release_date', ['primary_artist', 'name']],
                              meta_prefix='meta_')

# Show the flattened DataFrame
df_songs
```

```python
# Expample of a lyric
print(df_songs.iloc[0]['lyrics'])
```

```python
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation
    # text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.replace("\n"," ") # deleting new line sign
    text = text.replace("[pre-chorus]"," ") # deleting [pre-chorus]
    # Tokenize text
    return text
```

```python
lyrics_list = []
for idx in range(len(df_songs)):
    text_ = df_songs.loc[idx,'lyrics'].split("Verse 1]")[-1]
    text_ = preprocess_text(text_)
    lyrics_list.append(text_)
lyrics_list 
```

```python

# Pretend we have multiple songs in a list
lyrics = lyrics_list

# Tokenization and encoding sequences
vectorize_layer = tf.keras.layers.TextVectorization(
    max_tokens = 500,
    standardize = 'lower',
    output_sequence_length = 10,
    pad_to_max_tokens=True
)

vectorize_layer.adapt(lyrics)

sequences = vectorize_layer(lyrics)

# Create input sequences using list of  tokens - also short words
input_sequences = []
for song in sequences:
    for i in range(1, len(song)):
        n_gram_sequence = song[:i+1]
        input_sequences.append(n_gram_sequence)

# Pad sequences for equal length - each line the same
max_sequence_len = max([len(x) for x in input_sequences])

input_sequences = np.array(tf.keras.utils.pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre'))

# Predictors and label - 9 + 1 
predictors, label = input_sequences[:,:-1],input_sequences[:,-1]

label = tf.keras.utils.to_categorical(label, num_classes=vectorize_layer.vocabulary_size()+1) # changing labels to categoricla table
```

```python
predictors
```

```python
# designing neural network

model = tf.keras.Sequential()

model.add(tf.keras.layers.Embedding(
    input_dim=vectorize_layer.vocabulary_size()+1, output_dim=500, input_length=max_sequence_len-1)) # 1. layer 
    
model.add(tf.keras.layers.LSTM(150, return_sequences=True))

model.add(tf.keras.layers.Dropout(0.2))

model.add(tf.keras.layers.LSTM(100))

model.add(tf.keras.layers.Dense(vectorize_layer.vocabulary_size()+1, activation='softmax'))
 
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
# Load the TensorBoard notebook extension

%load_ext tensorboard

log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
history = model.fit(predictors, label, validation_split=0.2, epochs=400, batch_size=64, callbacks=[tensorboard_callback])
```

```python
def generate_text(seed_text, next_words, model, max_sequence_len):
    output_word = ""
    add_random = False
    for _ in range(next_words):
        token_list = vectorize_layer([seed_text])[0]
        if add_random:
            random_no_word = np.random.randint(vectorize_layer.vocabulary_size())
            # token_list.append()
            token_list = tf.concat([token_list[:-1], [random_no_word]],0)
            # print(token_list[:-1])
        token_list = [token for token in token_list if token != 0]
        token_list = tf.keras.utils.pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
        # print(token_list)

        predicted = model.predict(token_list, verbose=0)
        # print(np.argmax(predicted))
        word = vectorize_layer.get_vocabulary()[np.argmax(predicted)]
        
        if np.argmax(predicted) != 1 and word != output_word:
            output_word = word
            seed_text += " " + output_word
            add_random = False
            # print(seed_text)
        else: 
            add_random = True
    return seed_text
```
